## Question 8 >>

find all the countries in africa who drink more beer than average beer consupution in europe?

```python
import pandas as pd
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
# average beer consumption in Europe
europe_mean_beer_servings = drinks.loc[drinks['continent'] == 'Europe','beer_servings'].describe().mean()
```

```python
drinks.loc[(drinks['continent'] == 'Africa') & (drinks['beer_servings'] > europe_mean_beer_servings),:]
```

## Question 11 >

#### Question 11 a > Create a column which will have the total bill formed by multiplying quantity * item_price

#### Question 11 b > find the sum of all the total_bill for all items that contains chicken

```python
chips = pd.read_table('http://bit.ly/chiporders')
chips.head()
```

```python
chips['item_price'] = chips['item_price'].apply(lambda price: float(price.replace('$','')))
chips['total_bill'] = chips['quantity']*chips['item_price']
chips.head()
```

```python
chips_with_chicken_df = chips.loc[chips['item_name'].str.lower().str.contains('chicken'),:]
chips_with_chicken_df['total_bill'] = chips_with_chicken_df['quantity']*chips_with_chicken_df['item_price']
chips_with_chicken_df['total_bill'].sum()
```

```python
# 12 a > Create a column that has the first name of the passenger
# use the apply function

train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
train['first_name'] = train['Name'].apply(lambda name: name.split()[2].strip())
train.head()
```

```python
# 12 b > 
#    Create a column named "Survived_Text" where if the value in Survived column is 0 then "Survived_text" value is "Did not survive"
#    if the value in Survived column is 1 then "Survived_text" value is "Survived"

def get_survived_status(value):
    if value == 1:
        return ('Survived')
    elif value == 0:
        return ('Did not survive')
    else:
        return ('Data not provided')
```

```python
train['Survived_Text'] = train['Survived'].apply(get_survived_status)
train.head()
```

```python
# 12 c > 
    # select the columns First_Name, Sex, Age, Fare for those passengers
    # who travelled in first class and who did not survive
    
train.loc[(train['Pclass'] == 1) & (train['Survived_Text'] == 'Did not survive'), ['first_name', 'Sex', 'Age', 'Fare']]
```

```python
# 12 d > group the above dataset by sex and get the count of male and female passengers who did not survive
train.loc[train['Survived_Text'] == 'Did not survive',['Sex','Survived_Text']].groupby('Sex').count()
```
